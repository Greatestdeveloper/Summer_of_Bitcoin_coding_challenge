```python
import pandas as pd 

block_wt = 0 
block_fee = 0 
max_wt = 4000000
block_txns = []

txns_read = 0 
```

```python
df = pd.read_csv('mempool.csv') 
df.head(30) 
```

```python
# for i in range(len(sorted_df)) :
#     sorted_df['fee'][i] /= sorted_df['weight'][i]

# for i in range(len(sorted_df)) :
#     sorted_df['f2w_ratio'][i] = sorted_df['fee'][i] / sorted_df['weight'][i]

# sorted_df['f2w_ratio'] = sorted_df.apply( lambda row : row['fee'] / sorted_df['weight'], axis = 1 )


# df['f2w_ratio'] = round(float(df['fee']) / float(df['weight']), 10 )
for i in range(len(df)):
    df.at[i, 'f2w_ratio'] = round(float(df['fee'][i]) / float(df['weight'][i]), 6)


df.head(10)
```

```python
sorted_df = df.sort_values( by = 'f2w_ratio', ascending = False ) 
sorted_df = sorted_df.fillna('None') 
sorted_df.reset_index( inplace = True )
```

```python
# def allow_txn( id , block_txns ) : 
#     if( id == 'None' ) :
#         return( True ) 
#     ids = id.split(';') 
#     return( set(ids).issubset(set(block_txns)) ) # problem may lie here 

def allow_txn( id , block_txns ) : 
    if( id == 'None' ) :
        return( True ) 
    ids = id.split(';') 
    for parent in ids :
        if( parent not in block_txns ) : 
            return(False)
    return(True)
```

```python
for i in range(len(sorted_df)) : 
    print("i : ", i , " " , block_wt,"\n") 
    if( block_wt < max_wt ) : 
        txns_read += 1
        id = sorted_df.iloc[:,4][i]
        if( allow_txn(id , block_txns) and sorted_df['weight'][i] + block_wt <= max_wt ) : 
            block_wt += int(sorted_df['weight'][i])
            block_fee += int(sorted_df['fee'][i])
            block_txns.append( sorted_df['tx_id'][i])

# for i in range(len(sorted_df)) : 
#     print("i : ", i , " " , block_wt,"\n") 
#     id = sorted_df.iloc[:,4][i]
#     if( sorted_df['weight'][i] + block_wt <= max_wt and allow_txn(id)) : 
#         txns_read += 1
#         block_wt += sorted_df['weight'][i]
#         block_fee += sorted_df['fee'][i]
#         block_txns.append( sorted_df['tx_id'][i])
```
